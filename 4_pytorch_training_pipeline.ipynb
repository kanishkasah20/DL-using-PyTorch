# Numpy arrays to PyTorch tensors

using "from_numpy" function

```python
x_train_tensor = torch.from_numpy(x_train)
x_test_tensor = torch. from_numpy(x_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

# Defining the model

while defining the structure of neural network model we use OOPs in pytorch

```python
class MySimpleNN():

  # constructor
  def __init__(self, X):  #here X is my training set i.e x_train_tensor

     self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)  # weight will be assigned with random values & matrix size will be contains weights i.e no. of inputs we have (30,1)
     self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)                # bais is only one which is defined as zero in starting

  # forward pass
  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias   # z=wx+b
    y_pred = torch.sigmoid(z)                       # sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss

```

# Important Parameters

```python
learning_rate = 0.1  # lr is used while gradients descent or any optimizers
epoch = 25            # no. of runs or iterations of our data
```

# Training Pipeline

```python
# create model
model = MySimpleNN(x_train_tensor)

# define loop
for epoch in range(epoch):

  # forward pass
  y_pred = model.forward(x_train_tensor)
  # print(y_pred)

  # loss calculate
  loss = model.loss_function(y_pred, y_train_tensor)

  # backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradients     (before going to next epoch , we must need to make our parameters zero in pytorch, otherwise in pytorch weights or gradients values gets accumulated )
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

# Evaluation

```python
# model evaluation
with torch.no_grad():  # no_grad() beacuse during evaluation we are not calculating any gradients
  y_pred = model.forward(x_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
